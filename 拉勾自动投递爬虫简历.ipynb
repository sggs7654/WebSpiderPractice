```python
def load_cookies():
    with open("test.txt",'r+') as fr:
        cookies = eval(fr.read())
    return cookies

import selenium
from selenium import webdriver
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementNotInteractableException, ElementClickInterceptedException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time

def fast_get(browser,url):
    try:
        browser.get(url)
    except TimeoutException as e:
        return

def auto_click(get_fun, clue):
    try:
        button = get_fun(clue)
        button.click()
    except ElementNotInteractableException as e:
        time.sleep(0.5)
        auto_click(get_fun, clue)
    except (NoSuchElementException, ElementClickInterceptedException, RecursionError):
        pass
    return

options = webdriver.FirefoxOptions()
options.headless = False
browser = webdriver.Firefox(options=options)
browser.set_page_load_timeout(2)
#python爬虫 1-3年&不限
url = 'https://www.lagou.com/jobs/list_python%E7%88%AC%E8%99%AB?px=default&gj=3%E5%B9%B4%E5%8F%8A%E4%BB%A5%E4%B8%8B,%E4%B8%8D%E8%A6%81%E6%B1%82&city=%E5%85%A8%E5%9B%BD#filterBox'
#python数据 1-3年
# url = 'https://www.lagou.com/jobs/list_python%E6%95%B0%E6%8D%AE?px=default&gj=3%E5%B9%B4%E5%8F%8A%E4%BB%A5%E4%B8%8B&city=%E5%85%A8%E5%9B%BD#filterBox'

"""
覆盖Cookies
"""
fast_get(browser, "https://www.lagou.com")
browser.delete_all_cookies()
c = load_cookies()
for i in c:
    browser.add_cookie(i)


"""
自动投递
"""
fast_get(browser, url)
deliver_count = 0
last_page = 1
#开始投递循环
while True:
    elements = browser.find_elements_by_class_name('position_link')  # 获得岗位列表
    if deliver_count >= len(elements):  # 投完一页
        next_page_button = browser.find_element_by_class_name('pager_next ')
        next_page_button.click()  # 翻页
        page = browser.find_element_by_class_name('pager_is_current').text
        print("翻页，当前页：{}, 上一页：{}".format(page,last_page))
        if last_page == page:
            print("END")
            break
        last_page = page
        time.sleep(2)
        elements = browser.find_elements_by_class_name('position_link')  # 获得岗位列表
        deliver_count = 0
    elements[deliver_count].find_element_by_xpath("./h3").click()  # 在新标签页中进行简历投递
    handles = browser.window_handles
    browser.switch_to.window(handles[1])
    time.sleep(1.5)
    deliver_count += 1
    try:
        deliver_button = browser.find_element_by_class_name('resume-deliver')
        deliver_button.find_element_by_xpath(".//a[contains(text(), '已投递')]")
    except NoSuchElementException as e:  # 如果未投递
#         deliver_button = browser.find_element_by_class_name('resume-deliver')
        auto_click(browser.find_element_by_class_name, 'resume-deliver')
        time.sleep(0.1)
        auto_click(browser.find_element_by_class_name, 'resume-deliver')
        time.sleep(0.1)
        auto_click(browser.find_element_by_xpath, ".//a[contains(text(), '确 定')]")
        time.sleep(0.1)
        auto_click(browser.find_element_by_class_name, 'resume-deliver')
#         confirm_button = browser.find_element_by_id('delayConfirmDeliver')
        auto_click(browser.find_element_by_id, 'delayConfirmDeliver')
    browser.close()
    browser.switch_to.window(handles[0])
    

# browser.quit()
```
